```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
investments_df = pd.read_csv('crypto_daily_prices.csv')
investments_df
```

```python
investments_df.plot(x= 'Date', y =['BTC-USD Price', 'ETH-USD Price', 'LTC-USD Price'], label = ['Bitcoin Price', 'Ethereum Price', 'Litecoin Price'], linewidth = 3, figsize = (14,6))
plt.ylabel('Price[$]')
plt.title('My Crypto Prices!')
plt.legend(loc = "upper right")
plt.grid()
```

```python
daily_returns_df = pd.read_csv('crypto_daily_returns.csv')
daily_returns_df
```

```python
#Plot daily returns of BTC vs ETH
daily_returns_df.plot.scatter('BTC', 'ETH', color = 'r', figsize = (12, 7));
```

```python
import seaborn as sns
```

```python
#Plot daily returns of BTC vs ETH
plt.figure(figsize = (20, 10))
sns.scatterplot(daily_returns_df ['BTC'], daily_returns_df ['ETH']);
```

```python
# Define a dictionary with all crypto allocation in a portfolio
# Note that total summation = 100%

my_dict = {'allocation %' : [10, 10, 10, 60, 10]}
explode = (0, 0, 0, 0.2, 0)
```

```python
crypto_df = pd.DataFrame(data = my_dict, index = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'])
crypto_df.plot.pie(y = 'allocation %', explode = explode, figsize = (8,8))
plt.title('CRYPTO PORTFOLIO PIE CHART')
plt.legend(loc = "lower right")
```

```python
mu = daily_returns_df['ETH'].mean() # mean of distribution  
sigma = daily_returns_df['ETH'].std() # standard deviation of distribution

num_bins = 40

# Plot the histogram of the returns
plt.figure(figsize = (20, 15))

daily_returns_df['ETH'].plot.hist(bins = 30, alpha = 0.5, facecolor = 'r', figsize = (12,6))
plt.grid()
plt.ylabel('Probability')
plt.title('Histogram : mu=' + str(mu) + ' , sigma=' + str(sigma));
```

```python
plt.figure(figsize = (12,10))
cm = daily_returns_df.corr()
sns.heatmap(cm, annot = True)
```
